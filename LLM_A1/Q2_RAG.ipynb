```python
def generate_prompt(prompt):
    results = vector_store.similarity_search_by_vector(
        embedding=embeddings.embed_query(prompt), k=1
    )
    info = ""
    for doc in results:
        info = doc.page_content
        # print(f"* {doc.page_content} [{doc.metadata}]")
        return f"Knowledge: {info}, Query: {prompt}"

# prompt = "One side of the bottom layer of a triangular pyramid has 11 balls. How many are there in the whole pyramid? Note that the pyramid is equilateral and solid."
```

FACT CHECKING

```python
def outputs(query, rag_query):
  messages = [
      {"role": "user", "content": query},
  ]
  message_rag = [
      {"role": "user", "content": rag_query},
  ]
  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      messages,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.1,
      top_p=0.9,
  )
  outputs_rag = pipeline(
      message_rag,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.1,
      top_p=0.9,
  )
  o1,o2 = outputs[0]["generated_text"][-1]["content"], outputs_rag[0]["generated_text"][-1]["content"]
  return o1,o2
```

```python
def process_query(query):
    query_rag = generate_prompt(query)
    o1,o2 = outputs(query, query_rag)
    print("Output: ", o1)
    print("==============================================================================================================================")
    print(f"\nOutput with rag: {o2}")
```
