```python
def outputs(query, rag_query):
  messages = [
      {"role": "user", "content": query},
  ]
  message_rag = [
      {"role": "user", "content": rag_query},
  ]
  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      messages,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.1,
      top_p=0.9,
  )
  outputs_rag = pipeline(
      message_rag,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.1,
      top_p=0.9,
  )
  o1,o2 = outputs[0]["generated_text"][-1]["content"], outputs_rag[0]["generated_text"][-1]["content"]
  return o1,o2
```

```python
def process_query(query):
    query_rag = generate_prompt(query)
    o1,o2 = outputs(query, query_rag)
    print("Output: ", o1)
    print("\n================================================================================================================================================")
    print(f"\nOutput with rag: {o2}")
```

SELF CONSISTENCY

```python
def multiple_outputs(query, rag_query):
  messages = [
      # {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
      {"role": "user", "content": query},
      {"role": "user", "content": query},
  ]

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  message_rag = [
      {"role": "user", "content": rag_query},
  ]
  outputs = []

  # Loop over each message to query the model separately
  for message in messages:
      result = pipeline(
          [message],  # Process each message individually
          max_new_tokens=256,
          eos_token_id=terminators,
          do_sample=True,
          temperature=1.7,
          top_p=0.999,
      )
      # Assuming the model returns a list of dictionaries, append the response to outputs
      outputs.append(result[0]["generated_text"][-1]["content"])
  outputs_rag = pipeline(
      message_rag,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.1,
      top_p=0.9,
  )

  return outputs[0],outputs[1],outputs_rag[0]["generated_text"][-1]["content"]
```

```python
def consistency_check(query):
    query_rag = generate_prompt(query)
    print(query_rag)
    o1,o2, o3 = multiple_outputs(query, query_rag)
    print("Output 1: \n", o1)
    print("Output 2: \n", o2)
    print("\n================================================================================================================================================")
    print(f"\nRAG OUTPUT: {o3}")
```

EXAMPLE 1

```python
consistency_check("Which writer said, "" An atheist is a man who has no invisible means of support?")
```

EXAMPLE 2

```python
consistency_check("Which was the first European country to abolish capital punishment?")
```

EXAMPLE 3

```python
consistency_check("What famous comic strip character was inspired by the 1936 Henry Fonda film Trail of the Lonesome Pine?")
```
