```python
def generate_prompt(prompt):
    results = vector_store.similarity_search_by_vector(
        embedding=embeddings.embed_query(prompt), k=1
    )
    info = ""
    for doc in results:
        info = doc.page_content
        # print(f"* {doc.page_content} [{doc.metadata}]")
        return f"Knowledge: {info}, Query: {prompt}"

prompt = "Who was born first, Aaron Gillespie or Nathan Leone?"
print(generate_prompt(prompt))
```

```python

import torch
import warnings
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig

warnings.filterwarnings('ignore')
import transformers
import torch

model_id = "sarvamai/OpenHathi-7B-Hi-v0.1-Base"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

```python
query = "Which branch of medicine is concerned with disorders of the blood?"

# Generate output with the query
output = pipeline(query, max_length=100, do_sample=True, num_return_sequences=1, temperature=1.5)
o1 = pipeline(query, max_length=100, do_sample=True, num_return_sequences=1, temperature=1.5)
# Print the generated output
# print(output[0]["generated_text"])

# print(o1[0]["generated_text"])
def outputs(query, query_rag):
  o1 = pipeline(query, max_length=300, do_sample=True, num_return_sequences=1)[0]["generated_text"]
  o2 = pipeline(query_rag, max_length=300, do_sample=True, num_return_sequences=1)[0]["generated_text"]
  return o1,o2
```

FACT CHECKING

```python
def process_query(query):
    query_rag = generate_prompt(query)
    o1,o2 = outputs(query, query_rag)
    print("Output: ", o1)
    print("==============================================================================================================================")
    print(f"\nOutput with rag: {o2}")
```

SELF CONSISTENCY

```python
def multiple_outputs(query, rag_query):
  o1 = pipeline(query, max_length=300, do_sample=True, num_return_sequences=1, temperature = 1.5)[0]['generated_text']
  o2 = pipeline(query, max_length=300, do_sample=True, num_return_sequences=1, temperature = 1.6)[0]['generated_text']
  o3 = pipeline(rag_query, max_length=300, do_sample=True, num_return_sequences=1, temperature = 1.5)[0]['generated_text']
  return o1,o2,o3
```

```python
def consistency_check(query):
    query_rag = generate_prompt(query)
    o1,o2, o3 = multiple_outputs(query, query_rag)
    print("Output 1: \n", o1)
    print("Output 2: \n", o2)
    print("\n================================================================================================================================================")
    print(f"\nRAG OUTPUT: {o3}")
```

EXAMPLE 1

```python
consistency_check("Which branch of medicine is concerned with disorders of the blood?")
```

EXAMPLE 2

```python
consistency_check("How many husbands did the Wife of Bath have, as reported in Chaucer's Canterbury Tales?")
```

EXAMPLE 3

```python
consistency_check("By what pseudonym is writer Frederic Dannay Manfred Bennington Lee better known?")
```
