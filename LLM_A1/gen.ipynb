```python
!pip install pandas
import pandas as pd

# Load the uploaded movie dataset
file_path = 'movies.csv'
movies_df = pd.read_csv(file_path)


# Display the first few rows of the dataset to understand its structure
movies_df.head()
```

```python
# Function to generate a prompt for a movie
def generate_prompt(row):
    return f"Tell me about the movie {row['name']} released in {row['year']}. It is a {row['genre']} directed by {row['director']} and stars {row['star']}. What are some notable aspects of this movie?"

# Apply the function to each row in the dataframe to create a new column for prompts
movies_df['prompt'] = movies_df.apply(generate_prompt, axis=1)

# Display the first few rows with the generated prompts
movies_df[['name', 'prompt']].head()
```

```python
# Create a DataFrame for the classification task with 'genre' as the target variable
classification_df = movies_df.copy()

# Generate prompts for classification, excluding the 'genre' from the prompt
classification_df['classification_prompt'] = classification_df.apply(
    lambda row: f"Tell me about the movie {row['name']} released in {row['year']}. It is directed by {row['director']} and stars {row['star']}, duration of movie is {row['runtime']} minutes. What is Genre of this movie?",
    axis=1
)

# Create a DataFrame for the regression task with 'gross' as the target variable
regression_df = movies_df.copy()

# Generate prompts for regression, excluding the 'gross' from the prompt
regression_df['regression_prompt'] = regression_df.apply(
    lambda row: f"Tell me about the movie {row['name']} released in {row['year']}. It is directed by {row['director']} and stars {row['star']}, duration of movie is {row['runtime']} minutes. Can you predict the score of the movie?",
    axis=1
)

# Select only relevant columns for classification and regression
classification_df = classification_df[['classification_prompt', 'genre']]
regression_df = regression_df[['regression_prompt', 'score']]

# Display the first few rows of each DataFrame
classification_df.head(), regression_df.head()
```

```python
# Save the classification and regression DataFrames to separate CSV files
classification_df.to_csv('classification.csv', index=False)
regression_df.to_csv('regression.csv', index=False)
```
