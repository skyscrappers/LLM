```python
import pickle

with open('Llama_cot.pkl', 'rb') as f:
    llama = pickle.load(f)
with open('Phi_cot.pkl', 'rb') as f:
    phi = pickle.load(f)
with open('Gemma_cot.pkl', 'rb') as f:
    gemma = pickle.load(f)

print(gemma[0])
```

```python
with open("llama_cot.out", "w") as file:
    for ix, i in enumerate(llama): 
        line = i.split('\n')[-1] 
        file.write(f"{line}\n")

with open("phi_cot.out", "w") as file:
    for ix, i in enumerate(phi): 
        line = i.split('\n')[-1] 
        file.write(f"{line}\n")

with open("gemma_cot.out", "w") as file:
    for ix, i in enumerate(gemma): 
        line = i.split('\n')[0] 
        file.write(f"{line}\n")
```

```python
import re
with open('llama_cot.out', 'r') as f:
    text = f.read()

with open('phi_cot.out', 'r') as f1:
    text1 = f1.read()

with open('gemma_cot.out', 'r') as f2:
    text2 = f2.read()

print(text)
# llama_out = re.findall(r'[oO]ption\s+(\d)', text1)
# for i in llama_out:
#     print(i)

# with open('phi_response.pkl', 'wb') as f:
#     pickle.dump(llama_out,f)
```

```python
import re


options = []

pattern = re.compile(r'\boption (\d)\b', re.IGNORECASE)

responses = text.splitlines()

for response in responses:
    match = pattern.search(response)
    if match:
        options.append(f"{match.group(1)}")
    else:
        options.append('0')

for idx, option in enumerate(options):
    print(f"{option}")

with open('LLAMA_COT_RESPONSE.pkl', 'wb') as f:
    pickle.dump(options,f)
```
