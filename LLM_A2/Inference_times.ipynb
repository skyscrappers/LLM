```python
import transformers
import torch

llama_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline1 = transformers.pipeline(
    "text-generation",
    model=llama_id,
    model_kwargs={"torch_dtype": torch.bfloat16, "quantization_config" : {"load_in_4bit": True}},
    device_map="auto",

)

def llama_output(prompt, tokens):
    messages = [
        {"role": "user", "content": prompt},
    ]
#     print(messages)
    outputs = pipeline1(
        messages,
        max_new_tokens=tokens,
    )
    return outputs[0]["generated_text"][-1]["content"]

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
phi_id = "microsoft/Phi-3.5-mini-instruct"
pipeline2 = transformers.pipeline(
    "text-generation",
    model=phi_id,
    model_kwargs={"torch_dtype": torch.bfloat16, "quantization_config" : {"load_in_4bit": True}},
    device_map="auto",

)

def phi_output(prompt, tokens):
    messages = [
        {"role": "user", "content": prompt},
    ]
#     print(messages)
    outputs = pipeline2(
        messages,
        max_new_tokens=tokens,
    )
    return outputs[0]["generated_text"][-1]["content"]
gemma_id = "google/gemma-2b-it"
pipeline3 = transformers.pipeline(
    "text-generation",
    model=gemma_id,
    model_kwargs={"torch_dtype": torch.bfloat16, "quantization_config" : {"load_in_4bit": True}},
    device_map="auto",

)

def gemma_output(prompt, tokens):
    messages = [
        {"role": "user", "content": prompt},
    ]
#     print(messages)
    outputs = pipeline3(
        messages,
        max_new_tokens=tokens,
    )
    return outputs[0]["generated_text"][-1]["content"]
```

```python
zeroshot_prompts = []
cot_prompts = []
for i in range(100):
    zshot = f'''
Choose the answer to the given question from below options.
Question: {ds['test'][i]['question']}
option 1: {ds['test'][i]['choices'][0]}
option 2: {ds['test'][i]['choices'][1]}
option 3: {ds['test'][i]['choices'][2]}
option 4: {ds['test'][i]['choices'][3]}

Just output the correct option number like: option 1, option 2, option 3, option 4 without any explanation.'''
    cot= f'''
Choose the answer to the given question from below options.
Question: {ds['test'][i]['question']}
option 1: {ds['test'][i]['choices'][0]}
option 2: {ds['test'][i]['choices'][1]}
option 3: {ds['test'][i]['choices'][2]}
option 4: {ds['test'][i]['choices'][3]}

Explain step by step and then Output the correct option number like: option 1, option 2, option 3, option 4.'''
    zeroshot_prompts.append(zshot)
    cot_prompts.append(cot)
print(zeroshot_prompts[0])
print(cot_prompts[1])
```

```python
prompts = []
for i in range(100):
    prompt = f'''
Choose the answer to the given question from below options.
Question: {ds['test'][i]['question']}
option 1: {ds['test'][i]['choices'][0]}
option 2: {ds['test'][i]['choices'][1]}
option 3: {ds['test'][i]['choices'][2]}
option 4: {ds['test'][i]['choices'][3]}

Output the correct option number like: option 1, option 2, option 3, option 4.
'''
    prompts.append(prompt)
print(len(prompts))
```

```python
from tqdm import tqdm
import time
import pickle

c_llama, c_phi, c_gemma = [], [], []
z_llama, z_phi, z_gemma = [], [], []
cot = []
zeroshot = []
llama_times_cot, phi_times_cot, gemma_times_cot = [], [], []
llama_times_zeroshot, phi_times_zeroshot, gemma_times_zeroshot = [], [], []

for i in tqdm(range(10)):
    cot_prompt = cot_prompts[i]
    zeroshot_prompt = zeroshot_prompts[i]

    start_time = time.time()
    a1_cot = llama_output(cot_prompt, 1024)
    end_time = time.time()
    llama_times_cot.append(end_time - start_time)
    c_llama.append(a1_cot)

    start_time = time.time()
    a1_zeroshot = llama_output(zeroshot_prompt, 20)
    end_time = time.time()
    llama_times_zeroshot.append(end_time - start_time)
    z_llama.append(a1_zeroshot)

    start_time = time.time()
    a2_cot = phi_output(cot_prompt, 1024)
    end_time = time.time()
    phi_times_cot.append(end_time - start_time)
    c_phi.append(a2_cot)

    start_time = time.time()
    a2_zeroshot = phi_output(zeroshot_prompt, 20)
    end_time = time.time()
    phi_times_zeroshot.append(end_time - start_time)
    z_phi.append(a2_zeroshot)

    start_time = time.time()
    a3_cot = gemma_output(cot_prompt, 1024)
    end_time = time.time()
    gemma_times_cot.append(end_time - start_time)
    c_gemma.append(a3_cot)

    start_time = time.time()
    a3_zeroshot = gemma_output(zeroshot_prompt, 20)
    end_time = time.time()
    gemma_times_zeroshot.append(end_time - start_time)
    z_gemma.append(a3_zeroshot)

print(f"Average Llama Inference Time (COT Prompts): {sum(llama_times_cot) / len(llama_times_cot):.4f} seconds")
print(f"Average Llama Inference Time (Zeroshot Prompts): {sum(llama_times_zeroshot) / len(llama_times_zeroshot):.4f} seconds")
print(f"Average Phi Inference Time (COT Prompts): {sum(phi_times_cot) / len(phi_times_cot):.4f} seconds")
print(f"Average Phi Inference Time (Zeroshot Prompts): {sum(phi_times_zeroshot) / len(phi_times_zeroshot):.4f} seconds")
print(f"Average Gemma Inference Time (COT Prompts): {sum(gemma_times_cot) / len(gemma_times_cot):.4f} seconds")
print(f"Average Gemma Inference Time (Zeroshot Prompts): {sum(gemma_times_zeroshot) / len(gemma_times_zeroshot):.4f} seconds")

```

```python
from langchain_core.language_models.llms import LLM
import transformers
import torch
import pandas as pd
from pydantic import Field
from typing import Any, Dict
from langchain.agents import load_tools
from langchain.agents import initialize_agent

class CustomLLM(LLM):
    pipeline: Any
    def __init__(self, pipeline:Any):
        super().__init__(pipeline=pipeline)
        self.pipeline = pipeline
    def _call(self, prompt, **kwargs: Dict[str, Any]):
        messages = [
            {"role": "user", "content": prompt},
        ]
        outputs = self.pipeline(
            messages,
            max_new_tokens=512
        )
        return outputs[0]["generated_text"][-1]["content"]
    @property
    def _llm_type(self) -> str:
        """Get the type of language model used by this chat model. Used for logging purposes only."""
        return "CustomLLM"
```

```python
pipe = transformers.pipeline(
    "text-generation",
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    device_map="auto",
    model_kwargs={
    "torch_dtype": torch.bfloat16,
    "quantization_config": {"load_in_4bit": True},
    "low_cpu_mem_usage": True,
    }
)
pipe1 = transformers.pipeline(
    "text-generation",
    model="microsoft/Phi-3.5-mini-instruct",
    device_map="auto",
    model_kwargs={
    "torch_dtype": torch.bfloat16,
    "quantization_config": {"load_in_4bit": True},
    "low_cpu_mem_usage": True,
    }
)
pipe2 = transformers.pipeline(
    "text-generation",
    model="google/gemma-2b-it",
    device_map="auto",
    model_kwargs={
    "torch_dtype": torch.bfloat16,
    "quantization_config": {"load_in_4bit": True},
    "low_cpu_mem_usage": True,
    },
    generation_kwargs={
        "temperature": 0.7
    }
)
```

```python
from langchain import hub
from langchain.agents import AgentExecutor, load_tools
from langchain.agents.format_scratchpad import format_log_to_str
from langchain.tools.render import render_text_description
from langchain_community.utilities import SerpAPIWrapper
from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.llms import HuggingFacePipeline
import torch
from transformers import pipeline

device = 0 if torch.cuda.is_available() else -1

llm_llama = CustomLLM(pipeline=pipe)
llm_phi = CustomLLM(pipeline=pipe1)
llm_gemma = CustomLLM(pipeline=pipe2)
from langchain.agents.output_parsers.react_single_input import ReActSingleInputOutputParser
tools_llama = load_tools(["llm-math"], llm_llama)
tools_phi = load_tools(["llm-math"], llm_phi)
tools_gemma = load_tools(["llm-math"], llm_gemma)
agent_llama = initialize_agent(tools_llama, llm_llama, agent="zero-shot-react-description",
                         handle_parsing_errors = True,
                         agent_executor_kwargs={ "output_parser": ReActSingleInputOutputParser},
                         max_iterations=1, return_intermediate_steps=True)
agent_phi = initialize_agent(tools_phi, llm_phi, agent="zero-shot-react-description",
                         handle_parsing_errors = True,
                         agent_executor_kwargs={ "output_parser": ReActSingleInputOutputParser},
                         max_iterations=1, return_intermediate_steps=True)
agent_gemma = initialize_agent(tools_gemma, llm_gemma, agent="zero-shot-react-description",
                         handle_parsing_errors = True,
                         agent_executor_kwargs={ "output_parser": ReActSingleInputOutputParser},
                         max_iterations=1, return_intermediate_steps=True)
```

```python
from tqdm import tqdm
import time
import pickle

react_llama, react_phi, react_gemma = [], [], []
llama_times_react, phi_times_react, gemma_times_react = [], [], []

for i in tqdm(range(10)):
    prompt = prompts[i]

    start_time = time.time()
    try:
        out_llama = agent_llama({'input': prompt})
        if out_llama and 'intermediate_steps' in out_llama and len(out_llama['intermediate_steps']) > 0:
            react_llama.append(out_llama['intermediate_steps'][0][0].log)
        else:
            react_llama.append("NULL OUTPUT")
    except Exception as e:
        react_llama.append(f"Error: {str(e)}")
    end_time = time.time()
    llama_times_react.append(end_time - start_time)

    start_time = time.time()
    try:
        out_phi = agent_phi({'input': prompt})
        if out_phi and 'intermediate_steps' in out_phi and len(out_phi['intermediate_steps']) > 0:
            react_phi.append(out_phi['intermediate_steps'][0][0].log)
        else:
            react_phi.append("NULL OUTPUT")
    except Exception as e:
        react_phi.append(f"Error: {str(e)}")
    end_time = time.time()
    phi_times_react.append(end_time - start_time)

    start_time = time.time()
    try:
        out_gemma = agent_gemma({'input': prompt})
        if out_gemma and 'intermediate_steps' in out_gemma and len(out_gemma['intermediate_steps']) > 0:
            react_gemma.append(out_gemma['intermediate_steps'][0][0].log)
        else:
            react_gemma.append("NULL OUTPUT")
    except Exception as e:
        react_gemma.append(f"Error: {str(e)}")
    end_time = time.time()
    gemma_times_react.append(end_time - start_time)

with open('REACT_llama.pkl', 'wb') as f:
    pickle.dump(react_llama, f)
with open('REACT_phi.pkl', 'wb') as f:
    pickle.dump(react_phi, f)
with open('REACT_gemma.pkl', 'wb') as f:
    pickle.dump(react_gemma, f)

print(f"Average Llama Inference Time (ReAct Prompts): {sum(llama_times_react) / len(llama_times_react):.4f} seconds")
print(f"Average Phi Inference Time (ReAct Prompts): {sum(phi_times_react) / len(phi_times_react):.4f} seconds")
print(f"Average Gemma Inference Time (ReAct Prompts): {sum(gemma_times_react) / len(gemma_times_react):.4f} seconds")

print("ReAct time calculation process completed.")
```
